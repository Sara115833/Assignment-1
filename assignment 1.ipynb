```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
columns = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num',
    'marital-status', 'occupation', 'relationship', 'race',
    'sex', 'capital-gain', 'capital-loss', 'hours-per-week',
    'native-country', 'income'
]

data = pd.read_csv(data_url, header=None, names=columns, delimiter=',\s+', engine='python')
```

```python
data.head()
```

```python
data = data.replace('?', pd.NA)
data = data.fillna(data.mode().iloc[0])
```

```python
categorical_features = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']
data_encoded = pd.get_dummies(data, columns=categorical_features)
```

```python
features = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
X = data_encoded[features]
y = data_encoded['income']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)
```

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
confusion = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(confusion)
```

```python
tn, fp, fn, tp = confusion.ravel()
```

```python
sensitivity = tp / (tp + fn)
print("Sensitivity:", sensitivity)
```

```python
specificity = tn / (tn + fp)
print("Specificity:", specificity)
```

```python
y_prob = classifier.predict_proba(X_test)
pos_prob = y_prob[:, 1] 
```

```python
for prob in pos_prob:
    print("Posterior Probability of making over 50K a year:", prob)
```
